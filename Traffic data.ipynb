```python
days = ["Monday","Tuesday","Wednesday","Thursday","Friday"]

traffic_per_day = {}

for day, i in zip(days, range(0,135,27)):
    traffic_per_day[day] = traffic[i:i + 27]
```

```python
traffic_per_day["Monday"]
```

```python
for day in days:
    plt.plot(traffic_per_day[day]["Hour (Coded)"],traffic_per_day[day]["Slowness in traffic (%)"], label = day)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,12))


for day, i in zip(days, range(1,6)):
    plt.subplot(3,2,i)
    plt.plot(traffic_per_day[day]["Hour (Coded)"],traffic_per_day[day]["Slowness in traffic (%)"], label = day)
    plt.title(day)
    plt.ylim([0,25])

plt.subplot(3,2,6)

for day in days:
    plt.plot(traffic_per_day[day]["Hour (Coded)"],traffic_per_day[day]["Slowness in traffic (%)"], label = day)

plt.legend()
plt.show()
```
